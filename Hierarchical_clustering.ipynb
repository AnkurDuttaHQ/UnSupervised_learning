```python
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
data = load_iris()
data
X = data.data
y = data.target
X.shape
```

```python
## Visualize every feature with the first feature
fig ,ax = plt.subplots(2,2,figsize=(12,5))
sns.scatterplot(ax=ax[0][0],x=X[:,0],y=X[:,1])
sns.scatterplot(ax=ax[0][1],x=X[:,0],y=X[:,2],c=y)
sns.scatterplot(ax=ax[1][0],x=X[:,0],y=X[:,3])
```

```python
## Plotting the dendogram to see how agglomerative clustering is workung
from scipy.cluster.hierarchy import dendrogram , linkage
```

```python
# Feature scaling
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)

# Hierarchical clustering
z = linkage(x_scaled, method="ward")

# Plot dendrogram
plt.figure(figsize=(12,5))

dendrogram(z)

plt.xlabel("Samples or data points")
plt.ylabel("Distance between clusters")
plt.title("Dendrogram")

plt.tight_layout()
plt.show()
```

```python
## we train the algorithm on the k value of 2
clustering = AgglomerativeClustering(n_clusters=2)
label = clustering.fit_predict(x_scaled)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,2],hue=label)
```

```python
clustering = AgglomerativeClustering(n_clusters=3)
label = clustering.fit_predict(x_scaled)
sns.scatterplot(x=X[:,0],y=X[:,2],hue=label)
```
